```python
import numpy as np
import pandas as pd

def topsis(data, weights, impacts):
    norm_matrix = data / np.sqrt((data**2).sum(axis=0))
    weights = np.array(weights)

    # Reshape weights to be a row vector for broadcasting (1,number of columns in data)
    weights = weights.reshape(1, -1)

    weighted_matrix = norm_matrix * weights

    #best and worst
    best = np.where(np.array(impacts) == '+', weighted_matrix.max(axis=0), weighted_matrix.min(axis=0))
    worst = np.where(np.array(impacts) == '+', weighted_matrix.min(axis=0), weighted_matrix.max(axis=0))

    #distance
    dist_best = np.sqrt(((weighted_matrix - best) ** 2).sum(axis=1))
    dist_worst = np.sqrt(((weighted_matrix - worst) ** 2).sum(axis=1))

    scores = dist_worst / (dist_best + dist_worst)
    ranks = scores.argsort()[::-1] + 1

    return scores, ranks
```

```python
decision_matrix = data.iloc[:,1:].to_numpy()
```

```python
weights = [1, 1, 1, 1, 1,1,1]
impacts = ['+', '-', '-', '+','+','-','+']
```

```python
score,rank = topsis(decision_matrix, weights, impacts)
```

```python
data['Topsis_Score'] = score
data['Rank'] = rank
```
